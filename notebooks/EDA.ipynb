## Data Exploration and Feature Engineering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("../data/cost of living 2020.csv",index_col=0)
```

```python
df.drop("Unnamed: 9",axis=1,inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
def singleFeatureFromFile(filePath, df):
    file=open(filePath,'r')
    selected=[]
    for i in file.readlines():
        selected.append(i)
    selected="".join(selected)
    selected=selected.split("\n")
    feature=[1 if i in selected else 0 for i in df]
    return feature
```

The `../data/NoFees.txt` contains a list of countries that provide free Post Graduation programmes for Indian Students.

```python
df['Free Tuition']=singleFeatureFromFile("../data/NoFees.txt",df["Country"])
```

The `../data/AIBest.txt` contains a list of countries that provide the best Post Graduation programmes for Artificial Intelligence and Machine Learning.

```python
df['Good at AI']=singleFeatureFromFile("../data/AIBest.txt",df['Country'])
```

```python
df
```

```python
df=df[df['Cost of Living Index']<=100]
```

```python
sns.distplot(df['Cost of Living Index'])
```

```python
df['Cost of Living Index'].std()
```

```python
df['Avg Monthly Costs(USD)']=df['Cost of Living Index']*df['Cost of Living Index'].std()+df['Cost of Living Index'].mean()
```
